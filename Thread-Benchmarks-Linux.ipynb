```python
from matplotlib import pyplot as plt
import numpy as np

time_file = open("second_test")
lock_file = open("third_test")
new_results = open("new_results")
#new_file = open("thread_new_result")
times = time_file.readlines()
times_locks = lock_file.readlines()
#affinity_results = affinity_file.readlines()
new_time_results = new_results.readlines()

actual_times = [] 
lock_times = []
affinity_times = []
new_thread_times = []

for item in times:
    actual_times.append(int(item.split("\n")[0]))

for item in times_locks:
    lock_times.append(int(item.split("\n")[0]))

#for item in affinity_results:
#    affinity_times.append(int(item.split("\n")[0]))
    

for item in new_time_results[:-1]:
    new_thread_times.append(int(item.split("\n")[0]))
    
actual_times.sort()
lock_times.sort()
affinity_times.sort()
new_thread_times.sort()
print(actual_times)
#print(affinity_times)
print(new_thread_times)

```

```python
print(len(np.unique(actual_times)))
values, base = np.histogram(actual_times, 100)
cumulative = values.cumsum()


plt.title("Current pthread_create implementation")
plt.xlabel("Time taken by thread in microseconds")
plt.ylabel("Number of threads")
plt.plot(base[:-1], cumulative, c='blue')
#plt.hist(actual_times,50)
plt.show()


plt.title("Current pthread_create implementation")
plt.xlabel("Time taken by thread in microseconds")
plt.ylabel("Number of threads in each bin")
plt.hist(actual_times, 100)
plt.show()
```

```python
values, base = np.histogram(new_thread_times, 20)
cumulative = values.cumsum()

plt.title("New user thread library - sThreads")
plt.xlabel("Time taken by thread in microseconds")
plt.ylabel("Number of threads")
plt.plot(base[:-1], cumulative, c='blue')
#plt.hist(actual_times,50)
plt.show()

plt.title("New user thread library - sThreads")
plt.xlabel("Time taken by thread in microseconds")
plt.ylabel("Number of threads in each bin")
plt.hist(new_thread_times, 20)
plt.show()
```
